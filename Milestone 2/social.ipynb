```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```
social_media_interactions = pd.read_csv('M1social_media_interactions.csv')
social_media_interactions
```

```
# Visualizing the social media interactions data
plt.figure(figsize=(12, 6))
sns.barplot(data=social_media_interactions, x='Platform', y='Count', hue='Sentiment')
plt.title('Social Media Interactions by Platform and Sentiment')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.legend(title='Sentiment')
plt.show()
```
