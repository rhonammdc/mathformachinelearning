```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```
customer_transactions = pd.read_csv('M1 customer_transactions.csv')
customer_transactions
```

```
# Create a horizontal bar chart for product category with colors
plt.figure(figsize=(10, 6))
product_category_counts = customer_transactions['ProductCategory'].value_counts()
sns.barplot(x=product_category_counts, y=product_category_counts.index, palette='viridis')
plt.title('Number of Transactions per Product Category')
plt.xlabel('Number of Transactions')
plt.ylabel('Product Category')
plt.show()
```

```
# Create a horizontal bar chart for payment method with colors
plt.figure(figsize=(10, 6))
payment_method_counts = customer_transactions['PaymentMethod'].value_counts()
sns.barplot(x=payment_method_counts, y=payment_method_counts.index, palette='viridis')
plt.title('Number of Transactions per Payment Method')
plt.xlabel('Number of Transactions')
plt.ylabel('Payment Method')
plt.show()
```

The horizontal bar chart for payment methods has been created, displaying the number of transactions for each method using colors.

```
# Create a horizontal bar chart for transactions per quarter with colors
plt.figure(figsize=(10, 6))
quarter_counts = customer_transactions['Quarter'].value_counts()
sns.barplot(x=quarter_counts, y=quarter_counts.index, palette='viridis')
plt.title('Number of Transactions per Quarter')
plt.xlabel('Number of Transactions')
plt.ylabel('Quarter')
plt.show()
```

```
# Create a bell curve (normal distribution) plot for the 'Amount' column
plt.figure(figsize=(10, 6))
sns.histplot(customer_transactions['Amount'], kde=True, color='blue', bins=30)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```
