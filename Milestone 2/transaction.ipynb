```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```
customer_transactions = pd.read_csv('M1 customer_transactions.csv')
customer_transactions
```

```
# Visualizing the distribution of transactions across different product categories
plt.figure(figsize=(10, 6))
sns.countplot(data=customer_transactions, x='ProductCategory', order=customer_transactions['ProductCategory'].value_counts().index)
plt.title('Distribution of Transactions by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
# Visualizing the distribution of transactions across different payment methods
plt.figure(figsize=(10, 6))
sns.countplot(data=customer_transactions, x='PaymentMethod', order=customer_transactions['PaymentMethod'].value_counts().index)
plt.title('Distribution of Transactions by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
# Visualizing the distribution of transactions across different quarters
plt.figure(figsize=(10, 6))
sns.countplot(data=customer_transactions, x='Quarter', order=customer_transactions['Quarter'].value_counts().index)
plt.title('Distribution of Transactions by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
