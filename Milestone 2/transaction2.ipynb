> import pandas, matplotlib, seaborn, plotly

> load customer transactions

```
import pandas as pd

# Load the customer transactions data
customer_transactions = pd.read_csv('M1 customer_transactions.csv')
customer_transactions
```

> visualize use top product category based on sum of total amount, horizontal bar chart, use colors, sort from highest to lowest

```
import matplotlib.pyplot as plt
import seaborn as sns

# Group by ProductCategory and sum the Amount
category_amounts = customer_transactions.groupby('ProductCategory')['Amount'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=category_amounts, y=category_amounts.index, palette='viridis')
plt.title('Top Product Categories by Total Amount')
plt.xlabel('Total Amount')
plt.ylabel('Product Category')
plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the sum of total amount for each product category
category_amounts = customer_transactions.groupby('ProductCategory')['Amount'].sum().sort_values()

# Plot a horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=category_amounts, y=category_amounts.index, palette='viridis')
plt.title('Top Product Categories by Total Amount')
plt.xlabel('Total Amount')
plt.ylabel('Product Category')
plt.show()
```
